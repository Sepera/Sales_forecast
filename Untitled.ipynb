```python
%pylab inline
import pandas as pd
#import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pylab as plt
import warnings 
from itertools import product
from datetime import datetime
```

```python
df = pd.read_csv('Sales_data.csv', parse_dates= True, index_col= 'Date')
df = df.groupby([df.index])[['Sales']].sum().reset_index().set_index('Date')
```

```python
x = df['Sales']
plt.figure(figsize=(15, 6))
plt.plot(x)
plt.show()
```

```python
df['Sales_diff'] = df.Sales - df.Sales.shift(7) # сезонное дифференцирование
plt.figure(figsize(15,10))
sm.tsa.seasonal_decompose(df.Sales_diff.dropna()).plot()
print('Критерий Дикки-Фуллера: p = ', sm.tsa.stattools.adfuller(df.Sales_diff.dropna())[1])
```

```python
plt.figure(figsize(15,10))
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df.Sales_diff.dropna().values.squeeze(), lags = 35, ax=ax)
pylab.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df.Sales_diff.dropna().values.squeeze(), lags =35, ax=ax)
pylab.show()
```

```python
ps = range(0,3) 
d = 0
qs = range(0,3) # ок
Ps = range(0,4) # ok
D = 1
Qs = range(0, 4)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

```python
%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    # на некоторых параметрах модель не обучается
    try:
        model = sm.tsa.statespace.SARIMAX(x, 
                                          order = (param[0], d, param[1]), 
                                          seasonal_order= (param[2], D, param[3], 7)).fit(disp = -1)
    except:
        print('неправильный параметр', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
warnings.filterwarnings('default')       
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending = [True]).head())
```

```python
df['model'] = best_model.fittedvalues
plt.figure(figsize(15,7))
x[150:].plot()
df.model[150:].plot(color = 'r')
plt.ylabel('Sales')
pylab.show()
```
